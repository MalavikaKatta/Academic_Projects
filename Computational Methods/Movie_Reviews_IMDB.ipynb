```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup

# Load the CSV file into a DataFrame
df = pd.read_csv("RRR_Reviews.csv")

# IMDb URL for user reviews of the movie "RRR" (2022)
url = "https://www.imdb.com/title/tt8178634/reviews/?ref_=ttrt_sa_3"

# A list of film reviews
all_reviews = []
# The number of pages to scrape
num_pages = 400

for page in range(1, num_pages + 1):
    url = f"{url}&start={((page - 1) * 10)}"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        reviews = [review.get_text(strip=True) for review in soup.find_all("div", class_="text show-more__control")]
        all_reviews.extend(reviews)
    else:
        print(f"Failed to retrieve data from page {page}")
# Save all reviews to a CSV file
with open("RRR_Reviews.csv", "w", newline="", encoding="utf-8") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Review"])
    for review in all_reviews:
        writer.writerow([review])
print(f"Collected {len(all_reviews)} user reviews and saved to RRR_Reviews.csv")
df.head()
```

```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
import string
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from textblob import Word

# Previous code to access the RRR film reviews
url = "https://www.imdb.com/title/tt8178634/reviews/?ref_=ttrt_sa_3"
all_reviews = []
num_pages = 400
for page in range(1, num_pages + 1):
    url = f"{url}&start={((page - 1) * 10)}"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        reviews = [review.get_text(strip=True) for review in soup.find_all("div", class_="text show-more__control")]
        all_reviews.extend(reviews)
    else:
        print(f"Failed to retrieve data from page {page}")

# Defining a function to clean the text
def clean_text(text):
    # Remove special characters and punctuation
    text = ''.join([char for char in text if char not in string.punctuation])
    # Remove numbers
    text = ''.join([char for char in text if not char.isdigit()])
    # Lowercase text
    text = text.lower()
    # Tokenize text (split text into words)
    words = text.split()
    # Remove stopwords
    words = [word for word in words if word not in stopwords.words('english')]
    # Initialize a stemmer
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words]
    # Lemmatization
    words = [Word(word).lemmatize() for word in words]
    # Join the words back into text
    cleaned_text = ' '.join(words)
    return cleaned_text

# Clean and preprocess all reviews
all_reviews_cleaned = [clean_text(review) for review in all_reviews]
# Create a DataFrame with the original and cleaned reviews
data = {"Original_Review": all_reviews, "Cleaned_Review": all_reviews_cleaned}
df = pd.DataFrame(data)
# Save the DataFrame to a CSV file
df.to_csv("RRR_Reviews_Cleaned.csv", index=False)

print(f"Cleaned {len(all_reviews)} user reviews and saved to RRR_Reviews_Cleaned.csv")
df.head()
```

# **Question 3**

(30 points). Write a python program to conduct **syntax and structure analysis** of the clean text you just saved above. The syntax and structure analysis includes:

(1) Parts of Speech (POS) Tagging: Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) Constituency Parsing and Dependency Parsing: print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) Named Entity Recognition: Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
import spacy
from nltk import pos_tag, word_tokenize
from nltk.chunk import ne_chunk
from nltk.corpus import stopwords
from collections import Counter

# Load spaCy model
nlp = spacy.load("en_core_web_sm")
# Read the cleaned text from the CSV file
df = pd.read_csv("RRR_Reviews_Cleaned.csv")

# Function to perform Parts of Speech (POS) tagging
def pos_tagging(text):
    words = word_tokenize(text)
    tags = pos_tag(words)
    return tags

# Function to calculate the count of Nouns, Verbs, Adjectives, and Adverbs
def count_pos_tags(tags):
    noun_count = len([tag for word, tag in tags if tag.startswith('NN')])
    verb_count = len([tag for word, tag in tags if tag.startswith('VB')])
    adj_count = len([tag for word, tag in tags if tag.startswith('JJ')])
    adv_count = len([tag for word, tag in tags if tag.startswith('RB')])
    return noun_count, verb_count, adj_count, adv_count

# Function to perform Named Entity Recognition (NER) and count entities
def extract_entities(text):
    doc = nlp(text)
    entities = [ent.text for ent in doc.ents]
    entity_counts = dict(Counter(entities))
    return entity_counts

# Analyze each review
for i, row in df.iterrows():
    review = row['Cleaned_Review']
    # Perform POS tagging
    tags = pos_tagging(review)
    # Count POS tags
    noun_count, verb_count, adj_count, adv_count = count_pos_tags(tags)
    print(f"Review {i + 1} - POS Tag Counts: Noun={noun_count}, Verb={verb_count}, Adjective={adj_count}, Adverb={adv_count}")
    # Named Entity Recognition and entity count
    entity_counts = extract_entities(review)
    print(f"Review {i + 1} - Named Entities:")
    for entity, count in entity_counts.items():
        print(f"{entity}: {count}")

# Example: Constituency Parsing and Dependency Parsing
example_sentence = "The quick brown fox jumps over the lazy dog."
doc = nlp(example_sentence)
print("\nExample Sentence:")
print(example_sentence)
print("\nConstituency Parsing Tree:")
for token in doc:
    print(f"{token.text} ({token.dep_})")
print("\nDependency Parsing Tree:")
for chunk in doc.noun_chunks:
    print(f"{chunk.text} --> {chunk.root.text} ({chunk.root.dep_})")
```

**Constituency Parsing Tree:**
A constituency parsing tree, also referred to as a syntactic or phrase structure tree, reveals the hierarchical structure of a sentence by decomposing it into smaller constituents, which can be individual words or phrases (e.g., noun phrases or verb phrases). In the context of a constituency parsing tree:

*   The entire sentence is depicted by the tree's root.
*   Phrases or components are symbolized by non-terminal nodes within the  
    tree's structure.
*   Individual words or tokens are portrayed as terminal nodes in the tree.

**Dependency Parsing Tree:**
By constructing syntactic connections between words, a dependency parsing tree illustrates the grammatical structure of a sentence. It focuses on the relationships between words in relation to grammatical functions including subjects, objects, and modifiers. A dependency parsing tree contains:

*   A node in the tree represents each word in the sentence
*   Grammatical relationships are depicted by the edges (arcs) connecting nodes.
